```python
#tuples are immutable unlike list

# Why tuple?
# Clean, readable destructuring
# No temp variables

lst= list()
lst.append("a")
lst.append("b")
print(type(lst))
lst[0]="d"
print(lst)


tpl= ("a","b")
# tpl[0]="d" #immutable error 

print(type(tpl))
```

```python
#tuple mothods, count,index
alp = ("a","b","c","d","a")
print (alp.count("a")) #return's number of time the element exists
print (alp.index("d")) #return's index

#packing tuple
packed_tuple = 1,"Hello",2.3 # it will be treat a tuple
print(type(packed_tuple))

#unpacking tuple
a,b,c = packed_tuple
print (f"a:{a} b:{b} c:{c}")

```

```python
#nested tuple 

tp =  (("a","b"),("c","d"),"a","d")

print(tp[3])
print(tp[1][0])
```
