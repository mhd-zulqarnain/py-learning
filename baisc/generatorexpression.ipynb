```python
# Generator = values produced one at a time, only when needed
# returns generator and it uses ()

# Square numbers (single use)
gen = (x*x for x in range(5))
print(next(gen))
```

```python
# Iterating manually

gen = (x**x for x in [3,4,5,3])
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))

```
