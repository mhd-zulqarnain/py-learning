```python
# dictionaries are unordered collection of items,they store data in key-value pairs, keys must be unique and immutable

#define
obj = dict()
obj["name"]="test"

#to update object update
obj |= {
    "age":1,
    "iso":"pk"
}
obj.update({
    "class":"VIP"
})
#------


print(obj)
print(obj.values())
print(obj.keys())



```

```python
#iterataion in dictionaries

obj5 ={
     "age":2,
    "iso":"pk",
    "class":"0",
}

#to print keys
for item in obj5.keys():
    print(f"items {item}")

#to print values
for item in obj5.values():
    print(f"values {item}")

for key, value  in obj5.items():
    print(f"key {key} , value {value}")
```

```python
#copying mechanism in dictionaries
import copy

obj1 ={
     "age":2,
    "iso":"pk",
}

# print(obj1)

# copying work same like java i.e by reference called shallow copy
obj2= obj1
obj2["age"]=4
# print(obj1)

# deep copy works same as .copy like in kotlin i.e it makes a different copy
obj3= copy.deepcopy(obj1)
obj3["age"] =6
print(obj3)
print(obj1)


```

```python
# comprehension examples
#format [expression for in range condition]

num = [1,2,3,4,5,6]
squre = { x:x**2 for x in num}
# print(squre)


# swapping keys with values
tmp1 = {'name':'gtest','age': 6, 'iso': 'pk'} 
swap = { value:key for key,value in tmp1.items()}
# print(swap)


#Remove None / invalid values (VERY common)
tmp1 = {'name':'gtest',
        'age': 6, 
        'iso': 'pk',
        'class':"none",
        'cast':"ArithmeticError"
        } 

updatedObject = {
    key: value
    for key, value in tmp1.items() if value!="none"
}
# print(updatedObject)

# update the value exponentionaly
data = {
    "A": [1, 2],
    "B": [3, 4]
}

output= {
    (key,innerValue):innerValue**2
    for key, values in data.items()
    for innerValue in values
}

print(output)


```

```python
# return the active users
users = [
    {"id": 1, "name": "Ali", "active": True},
    {"id": 2, "name": "Sara", "active": False},
]

active= {
   u["id"]:u["name"]
   for u in users
   if u["active"]==True
}

print(active)
```
