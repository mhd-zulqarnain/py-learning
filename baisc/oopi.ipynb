```python
class User:
    def __init__(self ,name,age):
        self.__name= name #private access
        self.age = age  #public access
        
    def get_user_name(self):
        return self.__name
    
    # to avoid this error(error due to private property) on user.name    
    @property
    def name(self):
        return self.__name
        
user = User("test",1)
# user.__name   
print(user.get_user_name())
```

```python
#inheritance
class Car:
    def __init__(self,h_power,enginetype):
        self.h_power = h_power
        self.fuel = enginetype
        
    def fuel_type(self):
        print (f"the engine runs on {self.fuel} ")    
        
    def get_intro(self):
        print (f"the engine runs on {self.fuel} and it has {self.h_power} horse power ")
        
class EV(Car):
    def __init__(self, h_power, enginetype,is_selfdriving):
        super().__init__(h_power,enginetype) 
        self.is_selfdriving = is_selfdriving
        
    # method override
    def get_intro(self):
        super().get_intro()
        print(f"Self driving: {self.is_selfdriving}")
        

class Tmp(Car):
    def __init__(self, h_power, enginetype,had_four):
        super().__init__(h_power,enginetype) 
        self.had_four= had_four
        
    def get_intro(self):
        print(f"the wheel are powered: {self.had_four}")
    
 ## Function that demonstrates polymorphism
def get_intro(car):
    car.get_intro()
               
car = Car("1200","petro")  
tesla = EV("2000","electric",True)
jeep = Tmp("2000","electric","4x4")
get_intro(tesla)
get_intro(jeep)
```
