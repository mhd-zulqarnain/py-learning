```python
#lambda
#syntax 
#lambda arguments : expression

def sum1(a,b):
    return a+b
#lamba version 
sum = lambda a,b:a+b
squre = lambda a:a**2
isEven = lambda num: num%2==0
print(squre(3))
print(f"is 3 Even: {isEven(3)}")
print(f"is 6 Even: {isEven(6)}")
```

```python
#mapping in python
#syntax map(function:expresssion)
def squre1(a):
    return a**2
lst= [1,2,3,4,5,6]
map(squre1,lst)

## Lambda function with map
numbers=[1,2,3,4,5,6,7,8]
lam= lambda x:x*x
list(map(lam,numbers))

# print(output)
```

```python
#mapping with multiplr list
a =  [1,2,3,4,5]
b =  [6,7,8,9,0]

sum = lambda x,y:x+y 
result = list(map(sum,a,b))
print(result)
```

```python
a =  [1,2,3,4,5]
b =  [6,7,8,9,0]
c =  [6,7,8,9,0]

sum = lambda x,y,z:x+y+z
result = list(map(sum,a,b,c))
print(result)
```

```python
## map() to convert a list of strings to integers
str1= ["1","2","3"]

res= list(map(int,str1))
print(res)
```

```python
words = ['apple', 'banana', 'cherry']
upper_word = list(map(str.upper, words))
print(upper_word)
```

```python
#filter in python
#syntax filter(function:expresssion)

def even(a):
    return a%2==0

na =  [1,2,3,4,5,5,8,6,44,4,4,56,7,8,8,99,8645]

result =list( filter(even, na))
# print(result)

#filter the number greater than five in python

filtered = list(filter(lambda x:x>=5,na))
print(filtered)

```

```python
#filter in people wiht age greater tha 25
people = [
   { "name":"test1","age":45},
   { "name":"test2","age":46},
   { "name":"test3","age":11}
]

for person in people:
    if person["age"]>40:
        print(f"Aged {person["name"]}")
        
        
#using lamba
def isAged(person):
    return person["age"]>40

aged = list(filter(isAged , people))
print(aged)

     
    
        
```
