```python
#writing in a file overwriting
with  open("example.txt","w") as file:
    file.write("new line\nnew text")
        
```

```python
#writing in a file appending
with open("example.txt","a") as file:
    file.write("\nappended text") 
```

```python
#writing multiple lines
lines=['First line \n','Second line \n','Third line\n']
with open("example.txt","w") as file:
    file.writelines(lines)
```

```python
### Binary Files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open("example.bin","wb") as file:
    file.write(data)
    
with open("example.bin","rb") as file:
   content=  file.read()
   print(content)
```

```python
### Read the content froma  source text fiile and write to a destination text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
# ðŸ”€ Combined modes 
# Mode	Meaning
# "w"	write (overwrite)
# "w+"	write + read
# "w+b"	write + read (binary)
# "wt"	write (text)
# "wb"	write (binary)
# "wb+"	write + read (binary)
# "wt+"	write + read (text)


# writing in and reading file at the same time
with open("example.txt","a+") as file:
    file.write("\nnew changes")
    file.seek(0) ## Move the file cursor to the beginning
    content = file.read()
    print(content)
```

```python
with open("example.txt","r") as file:
    line = file.readline()
    char_count = sum(len(x) for x in line)
    word_count = sum(len(x.split()) for x in line)
    print(char_count)
    print(word_count)
```

```python
#exception handling
#try except else finally 

try:
    inpt = int(input("Enter a number"))
    result = 43/inpt
except ValueError as ex:
    print(f"only the numbers are excepted :{ex} ")
except ZeroDivisionError:
    print("the number should be greater than 0")
else: 
    print("the output is ",result)
finally :
     print("execution completed ")
    
```
