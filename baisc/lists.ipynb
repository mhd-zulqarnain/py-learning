```python
#list comprehension 
# basic syntax          [experession for item in iterable]
# with condition        [experession for item in iterable if condition]
# nested list           [ experession for item in iterable for item in iterable]
# with condition else   [expression_if_true if condition else expression_if_false for item in iterable]



squre= [num**2 for num in range(10)]
print(f" basic syntax {squre}")

squreOfOdd = [num**2 for num in range(11) if num%2!=0]
print(f" with condition {squreOfOdd}")

squreOfNumberLessThanFive = [num**2 if num<5 else num for num in range(11)]
print(f" with condition else {squreOfNumberLessThanFive}")

alphabat =["a","b","c"]
nestedList = [(i,j) for i in range(3) for j in alphabat ]
print(f" with condition {nestedList}")

words = ["ArithmeticError","AttributeError","BufferError","ConnectionError"]

sizeOfWords = [ len(word) for word in words]
print(f" size Of Words {sizeOfWords}")
```

```python
#slicing list
# Positive index →   0       1       2       3       4
#                    |       |       |       |       |
# List values     ["word 1","word 2","word 3","word 4","word 5"]
#                    |       |       |       |       |
# Negative index →  -5     -4      -3      -2      -1
# list[start : end: step]


list = [1,2,3,4,5,6,7,8,9]
print(list[0:5:2])

print(list[::2]) # move +2 each time
print(list[::-2]) # reverse move +2 each time

print(f"start from -3 index to end {list[-3:]}") 
print(f"start from -3(6 excluded) index to end at -8(2) {list[-8:-4]}") 

```

```python
#check the feed back of users are negative or positive
# with condition        [experession for item in iterable if condition]


feedback =["Great service","Very satisfied","poor service","Great experience"] 
feedback.append("not satified")

postive_feedback_count = [1 for comment in feedback if  "good" in comment.lower() or "great" in comment.lower() ]
negative_feedback_count = [1 for comment in feedback 
                           if "good" not in comment.lower() 
                           or "great" not in comment.lower()]

# string format {value_if_true if condition else value_if_false}

formated  = f"the feeedback is {'positive' if sum(postive_feedback_count)>sum(negative_feedback_count)  else 'negative'}"
print(formated)
```
