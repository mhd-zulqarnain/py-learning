```python
#slicing list
# Positive index →   0       1       2       3       4
#                    |       |       |       |       |
# List values     ["word 1","word 2","word 3","word 4","word 5"]
#                    |       |       |       |       |
# Negative index →  -5     -4      -3      -2      -1
# list[start : end: step]


list = [1,2,3,4,5,6,7,8,9]
print(list[0:5:2])

print(list[::2]) # move +2 each time
print(list[::-2]) # reverse move +2 each time

print(f"start from -3 index to end {list[-3:]}") 
print(f"start from -3(6 excluded) index to end at -8(2) {list[-8:-4]}") 

```

```python
#in build methods for list
list = ["word 1","word 2","word 3","word 4","word 5"]

newWord ="word 6"
list.append(newWord)
list.insert(1,newWord)
# list.remove(1) this won't work we need to use list.pop(1) or list.remove(newWord)
print(list)
print(list.count(newWord))
# list.remove(newWord) will remove all newWord

#list.reverse()
list.sort()
print(list)
```
