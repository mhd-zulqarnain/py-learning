```python
#poistional or varg parameter in kotlin

def printAnthing(*params):
    for i in params:
        print(i)
    
printAnthing("k",[2,3,3,4],{"name":"test"},(23,4,5,6))
```

```python
#keywod argument

# Kotlin mental mapping
# Python	Kotlin
# *args	    vararg
# **kwargs	Map<String, Any>

def printAnthing1(**args):
    for key,value in args.items():
        print(f" {key} {value} ")
    
printAnthing1(name="Ali", age=25) #Python does this automatically {"name": "Ali", "age": 25}
```

```python
#check plandrom 

def isPalindrom(word):
     tmp = word.lower().replace(" ","")
     return tmp == tmp[::-1]

word  = "Hello"
word1  = "A man a plan a canal Panama"

print(f"{word}-> palindrom status:{ isPalindrom(word)}")
print(f"{word1}-> palindrom status: {isPalindrom(word1)}")


    
```

```python
# Recuression to fin factorial
# 5! = 5x3x4x2x1

def factorial(number):
    if number==0:
        return 1
    else:
        return number * factorial(number-1)
   
      
number = 4 
print(f"factorial of {number} is {factorial(number)} ")
    
```
