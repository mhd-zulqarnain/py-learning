````python
#calculor using if statements

num1 = float(input("Enter first number"))
num2 = float(input("Enter second number"))

operator = input("Enter the operator x,+,-,/,xx")
if operator == "+":
    output = num1 + num2
elif operator == "-":
    output = num1 - num2
elif operator == "x":
    output = num1 * num2
elif operator == "xx":
    output = num1 ** num2
elif operator == "/":
    if num2 != 0:
        output = num1 / num2
    else:
        output = "invalid input"
else:
    output = "invalid input"

print("the output is", output)  
````
