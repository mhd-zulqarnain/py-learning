```python
### Tokenization 
## Paragraph-->words 
import nltk

# Download required tokenizer data
# nltk.download('punkt')
# nltk.download('punkt_tab')

corpus="""Hello Welcome,to NLP Tutorials.
Please do watch the course! to become expert in NLP.
"""

from nltk.tokenize import sent_tokenize,word_tokenize
document = sent_tokenize(corpus)
# print(document)

#word tokenization 
wordTokenization = nltk.word_tokenize(corpus)
# print(wordTokenization)

# tokenization of sentence
for word in document:
    print(nltk.word_tokenize(word))
```

## Stemming
Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

```python
## Classification Problem
## Comments of product is a positive review or negative review
## Reviews----> eating, eat,eaten [going,gone,goes]--->go

import nltk 
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]

stemming = nltk.stem.PorterStemmer()
for word in words :
    print(stemming.stem(word))
    
    
#using regex
reg_stemmer=nltk.stem.RegexpStemmer('ing$|s$|e$|able$', min=4)
print(f"--using regex{reg_stemmer.stem('eating')}")

# Snowball Stemmer (better than portstemmer)
snowball_stemmer = nltk.stem.SnowballStemmer("english")
print(f"--------")

for word in words :
    print(f"{word}-->{snowball_stemmer.stem(word)}" )
```
